**4) Plotting a Waveform**

```python
import matplotlib.pyplot as plt
import librosa.display

x, sr = librosa.load('/content/sample_data/english.wav')
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)
```

**5) Convert Voice Recordings to Text**

**5.1) English.wav**

```python
import speech_recognition as sr
s2t = sr.Recognizer()
with sr.AudioFile('/content/sample_data/english.wav') as source:
  sound2text = s2t.listen(source)
text = s2t.recognize_google(sound2text)
print(text)
```
